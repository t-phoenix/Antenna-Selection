```python
def location(number, m_column):
    return x_co(number,m_column), y_co(number,m_column)

def x_co(number, m_column):
    if number%m_column ==0 :
        x = m_column-1
    else: 
        x = (number%m_column)-1
    return x

def y_co(number,m_column):
    if number%m_column ==0:
        y = int(number/m_column) -1
    else:
        y = int(number/m_column)
    return y


#a= [[0]*m_column for i in range(m_row)]
location_list=[]
temp_l=[]
for elem in l:
    for i in elem:
        temp_l.append(location(i,m_column))
    location_list.append(temp_l)
    temp_l=[]
        


    
for elem in location_list:
    print(elem)
```

```python
def inner_prod(list1,list2):
    new_list =[]
    new_list= np.array(list1)*np.array(list2)
    total=0
    for ele in range(0, len(new_list)): 
        total = total + new_list[ele]
    return total    
    
```

```python
def steering_vector(phi,list): #pass antenna_location in list
    theta = math.radians(0)
    sv_list=[]
    wavelength = 3/520
    k0 = 2*math.pi/wavelength
    k = [math.cos(theta)*math.sin(phi),math.sin(theta)]
    
    for i in list: 
        n=np.dot(k,i)
        sv_list.append(cmath.exp(complex(0,-k0*n)))
    return sv_list    
  
#sv0=steering_vector(0,antenna_location)  
                                        # but phi vary from -pi to pi
#for ele in sv0:
#    print(ele)
```

```python
def weight_vector(selection_list,array_list):#list --> the antenna selected for k user
    theta_k = math.radians(0)
    phi_k = math.radians(0)
    
    wavelength = 3/520
    k0 = 2*math.pi/wavelength
    M=len(selection_list)
    k = [math.cos(theta_k)*math.sin(phi_k),math.cos(theta_k)]
    
    for i in range(0,len(array_list)):
        if array_list[i] in selection_list:
            n=np.dot(k,location(array_list[i],m_column))
            array_list[i]=cmath.exp(complex(0,k0*n))/M
            
        else:
            array_list[i]=0
        
    return array_list    


temp=[]
temp = list(antenna_array)


wv=[]
wv=weight_vector(temp,temp)                                    # selection of antenna can be chose

for ele in wv:
    print(ele)
```
