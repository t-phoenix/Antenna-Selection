```python
# beam_magnitude
def beam_magnitude(list,phi):#include theta and phi as arguements to steer
    wavelength = 3/520
    theta_k=math.radians(0)
    phi_k=math.radians(0)
    theta = 0
    #phi = math.pi/4 temp change the anglesz
    uky=math.cos(theta_k)*math.sin(phi_k) #angles in form of pi->> math.pi/2 or math.pi/3
    ukz=math.sin(theta_k)
    uy=math.cos(theta)*math.sin(phi)
    uz=math.sin(theta)
    k0=2*math.pi/wavelength
    terms =[]
    bw=0
    for elem in list:
        terms.append(cmath.exp(complex(0,k0*(wavelength/2)*((elem[0]*(uky-uy))+(elem[1]*(ukz-uz))))))
        
    for i in terms:
        bw = bw+i
        
    return abs(bw)    

    

```

```python
def beamwidth_degrees(beamlist):
    beam_req=max(beamlist)/math.sqrt(2)
    #half_bw=0
    count=0
    for i in beamlist:
        if i<beam_req:
            #half_bw=i
            break
        count=count+1 
    bw= math.degrees(2*(phi_list[count]))
    count=0
    return bw

```

```python
def null_point(list1): #list1 contains beam_list (obtained by beam_magnitude)
    current=list1[0]
    count=0
    for i in list1:
        if i > current:
            break
        elif i < current:
            current = i
            count=count+1
    return count

#null_point(beam_list) 
            
```

```python
def side_lobe(test_list ,phi_list): # true (1)->> there is a side lobe
    threshold=max(test_list)/2 # 50%
    null_index=null_point(test_list)
    value=-1
    for i in range(null_index,len(phi_list)):
            if test_list[i]>threshold:
                value =1
                break
            else:
                value = 0
    return value            
```

```python
beamwidth_list=[] # beamwidths for all combinations
new_selections=[]
for selection in location_list:
    temp_beam_list=[]
    for phi_i in phi_list:
        temp_beam_list.append(beam_magnitude(selection,phi_i))
    if side_lobe(temp_beam_list,phi_list) == 0:    
        beamwidth_list.append(beamwidth_degrees(temp_beam_list))
        new_selections.append(selection)
    else:
        temp_beam_list=[]
```
