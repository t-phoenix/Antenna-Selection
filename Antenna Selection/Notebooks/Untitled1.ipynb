```python
af_list=[]
phi_list=[]
phi= -math.radians(180)
theta=math.radians(0)
while phi<math.radians(180):
    phi_list.append(phi)
    temp_sum=0
    for a in range(0,9):
        for b in range(0,9):
            temp_sum=temp_sum+(cmath.exp(complex(0,(-math.pi*(a*math.cos(theta)*math.sin(phi)+b*math.sin(theta))))))
    af_list.append(abs(temp_sum))
    temp_sum=0
    phi=phi+math.radians(5)        
```
